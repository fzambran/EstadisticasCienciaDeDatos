<a id="tipos-de-variables"></a>
### Tipos de variables
* **Cualitativas nominales**: Hace referencia a variables en las cuales sus categorías no representan una jerarquía. NO ES POSIBLE ORDENARLAS. Ejemplos: género, grupo sanguíneo, regiones de chile, religión, ideología, nacionalidad, etc.
* **Cualitativas ordinales**: Por el contrario de las nominales, sus categorías se pueden ordenar en forma jerárquica. Ejemplos: calificación crediticia, calidad de la atención, etapas de la vida, meses del año, etc.
* **Cuantitativas discretas**: Son aquellas variables numéricas que solamente toman determinados valores y no cualquier valor que se encuentre entre dos de ellos. Ejemplos: cantidad de asistentes a un workshop, número de árboles que hay en un parque, cantidad de dedos en la mano, etc.
* **Cuantitativas continuas**: Son aquellas que cuando entre dos valores cualquiera, pueden existir valores intermedios. Ejemplos: edad, estatura, peso, distancias, velocidad, tiempo de demora, etc.

<a id="que-es-un-grafico"></a>
### ¿Qué es un gráfico?
<img src="https://estebanrucan.github.io/personal-book/figuras/barras.jpeg" width="500">

* Es una representación visual de datos estadísticos, de manera tal que podamos interpretarlos, analizarlos y entenderlos de forma más sencilla.
* Realizamos gráficos cuando deseamos realizar un **análisis exploratorio de datos**, por ejemplo, para descubrir el comportamiento de una variable.
* También cuando necesitamos __exhibir__ los resultados de un análisis al __público__, en este caso prestamos más atención a la estética.
* *Características de una buena visualización de datos*: agradable a la vista, funcional, muestra hallazgos y basada en datos confiables, o sea que transmite la verdad.
* Una imagen vale más que mil palabras.

### Tipos de gráficos
* El gráfico que seleccionaremos depende de las variables que nos interese visualizar.
* No siempre es necesario usar un gráfico, a veces una tabla es más que suficiente.

* En la función `plot`, debemos agregar el argumento `kind="tipo"`.

<a id="grafico-de-torta"></a>
### Gráfico de Torta
* Muestra la distribución de las frecuencias a través de dividir el área total de un círculo en partes que representen la frecuencia de cada categoría  en una variable cualitativa.
* No es recomendable usar gráficos de torta cuando:
  * Tengamos más de 5 categorías.
  * Las categorías tienen frecuencias muy similares.
  * Hay categorías con frecuencias demasiado pequeñas.
* Debemos utilizar la función `plt.pie()`, indicando los valores a representar y sus etiquetas respectivas.

````python
# Ejemplo de gráfico de torta, Cantidad de productos vendidos por categoría
data_pie = data.groupby(["Category"])["Quantity"].sum()
print(data_pie)

plt.pie(data_pie, labels=data_pie.index, autopct="%1.1f%%");
````

<a id="grafico-de-puntos"></a>
### Gráfico de Puntos
* Conocido también como scatter plot, gráfico de dispersión o dispersograma. Gráfica cada par de valores en dos variables.
* Esta visualización se usa para encontrar relaciones y patrones existentes entre dos variables cuantitativas.
* Algunas relaciones comunes:
  * Lineal, logarítmica, exponencial, etc
  * Fuerte o débil.
  * Positiva o negativa.
* Evitar usar este gráfico hay muchos datos, porque las relaciones suelen hacerse menos evidentes. En este caso es mejor usar un mapa de calor.
* Debemos agregar el argumento `kind="scatter"`.
* También podemos agregar parámetros como `c`, que hace referencia al color de los puntos, puede ser exactamente un color u otra variable.
* Si estamos coloreando por una variable numérica, podemos especificar una paleta de colores con el parámetro `cmap`. Paletas usuales son `Blues`, `viridis`, `magma` e `inferno`.

````python
# Ejemplo 2: Al gráfico de puntos realizado inicialmente, coloree por la variable Discount
data.plot(
    x    = "Profit CLP",
    y    = "Sales CLP",
    kind = "scatter",
    c    = "Discount"
);
````

<a id="grafico-de-lineas"></a>
### Gráfico de Líneas

* También conocido como gráfico de tendencias.
* Se muestra a través de una serie de puntos conectados por una línea.
* Generalmente es usado para analizar comportamientos de una variable cuantitativa dentro de un intervalo de tiempo.
* Es el tipo de gráfico usado por defecto, pero si deseamos ser específicos agregamos `kind="line"`.
* Si deseamos añadir los puntos, agregar `marker="."`.
* Si necesitamos un gráfico de líneas agrupado, podemos utilizar `seaborn` y su función `lineplot()`.

````python
# Ejemplo: Análizar las tendencias de cantidad de ventas por día de la semana según categoría.
orden = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

data["Order Weekday"] = pd.Categorical(data["Order Weekday"], categories=orden, ordered=True)

sns.lineplot(
    x         = "Order Weekday",
    y         = "Quantity",
    hue       = "Category",
    data      = data,
    estimator = sum,
    errorbar  = None,
    marker    = "o"
);
````

<a id="aspectos-esteticos-en-matplotlib"></a>
### Aspectos estéticos en matplotlib
* En `matplotlib` existen muchas posibilidades para customizar gráficos.
* En todos nuestros gráficos debemos procurar tener las etiquetas correctas, esto incluye la leyenda apropiada.
* Estas deben explicar la mayor cantidad de información posible sin perder simplicidad.
* Por ejemplo en un gráfico de puntos, debemos tener ambos ejes etiquetados y un título acorde.
* La sintaxis apropiada para un gráfico debe ser definiendo inicialmente el panel donde irá la información.
* Para ello usamos la función `subplots` otorgada por el módulo `pyplot`.
* Un argumento básico de `subplots` es `figsize`, donde debemos indicar una tupla con el ancho y alto respectivamente,
  
```python
fig, ax = plt.subplots(figsize=(10, 8))
data.plot(..., ax=ax)
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_title("Título")
```

````python
# Ejemplo: agregar etiquetas al gráfico de puntos.
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8)) # Hoja en blanco
data.plot(
    x    = "Profit CLP",
    y    = "Sales CLP",
    kind = "scatter",
    c    = "Discount",
    cmap = "viridis",
    ax   = ax # poner gráfico en la hoja en blanco
)
ax.set_xlabel("Ganancia (CLP)")
ax.set_ylabel("Ventas (CLP)")
ax.set_title("Ganancias y Ventas por Descuento (%)");
````

<a id="formato-de-las-marcas-de-eje"></a>
### Formato de las Marcas de Eje

* Para una correcta comprensión de los gráficos, debemos procurar que la graduación de los ejes sea lo más sencilla de entender posible.
* Podemos establecer los valores que se visualizarán en las marcas de ejes, y a su vez, de que forma mostrarlos.
* Para definir las marcas de ejes a mostrar, utilizamos `ax.set_xticks()` o `ax.set_yticks()` según corresponda. Le debemos entregar una lista de valores numéricos.
* Para establecer la forma en que se mostrarán las marcas de eje, se usa `ax.set_xticklabels()` y `ax.set_yticklabels()` según corresponda. Se puede usar una lista por comprensión para recorrer la lista de valores numéricos.

```python
fig, ax = plt.subplots(...)
data.plot(..., ax=ax)
ax.set_xticks(valores_x)
ax.set_yticks(valores_y)
ax.set_xticklabels([f(x) for x in valores_x])
ax.set_yticklabels([f(y) for y in valores_y])
```

````python
# Ejemplo: modificar graduación de ejes al gráfico de puntos.

fig, ax = plt.subplots(figsize=(10, 8)) # Hoja en blanco
data.plot(
    x    = "Profit CLP",
    y    = "Sales CLP",
    kind = "scatter",
    c    = "Discount",
    cmap = "viridis",
    ax   = ax # poner gráfico en la hoja en blanco
)
# Etiquetas
ax.set_xlabel("Ganancia (CLP)")
ax.set_ylabel("Ventas (CLP)")
ax.set_title("Ganancias y Ventas por Descuento (%)")
# Formato de eje
f = lambda x: f"${x:,.0f}" if x >= 0 else f"-${-x:,.0f}"
rango_x = range(-5000000, 7000001, 2000000)
rango_y = range(0, 18000001, 2500000)
label_x = [f(x) for x in rango_x]
label_y = [f(y) for y in rango_y]
ax.set_xticks(rango_x)
ax.set_yticks(rango_y)
ax.set_xticklabels(label_x)
ax.set_yticklabels(label_y);
````

### Panel con múltiples gráficos

* En muchas situaciones, es conveniente dejar múltiples gráficos que estén asociados entre sí en un solo panel.
* **Por ejemplo:** podríamos dejar en un solo panel un histograma y un *boxplot*.
* Para estos propósitos, podemos configurar múltiples gráficos, que quedarán en un solo panel mediante los parámetros `ncols` y `nrows` de la función `plt.subplots()`.
* Estudiemos más de cerca la destructuración que realizamos cuando usamos la función `plt.subplots()`

````python
resultado = plt.subplots()
print(resultado)
````

Como podrán ver, la función `plt.subplots()` entrega una tupla con las instancias de dos clases: `Figure` y `Axes`, por lo que al momento de realizar la destructuración, obtenermos en la variable `fig` la instancia de `Figure` y  en `ax` la instancia de `Axes`.

````python
fig, ax = plt.subplots()
print("- fig:", fig)
print("- ax :", ax)
````

Ahora veamos el resultado cuando necesitamos graficar paneles con múltiples gráficos. Por ejemplo, realicemos un panel que nos permita agregar dos gráficos en una columna.

````python
resultado = plt.subplots(nrows=2, ncols=1)
print(resultado)
````

* Si bien en primer lugar obtenemos una instancia de `Figure`, en el segundo lugar de la tupla ahora obtenemos un `array` de *Numpy*: en cada una de sus posiciones entrega una instancia de `Axes`.
* Entonces para realizar un panel con dos gráficos tenemos que considerar a través de la variable `ax` la posición de cada uno de estos.
* Veamos un ejemplo:

````python
# Panel en blanco
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 4))
print("-ax:", ax)
# Gráfico 1
ax[0].pie(x=data_pie, autopct="%1.1f%%", labels=data_pie.index)
# Gráfico 2
data_barplot.plot(kind="barh", ax=ax[1])

# Etiquetas
fig.suptitle("Comportamiento de las ventas en las categorías y segmentos") # Título del panel
ax[1].set_xlabel("Ventas (CLP)");
#ax[0].set_title("Gráfico de Torta")
````
